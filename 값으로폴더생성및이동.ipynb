```python
import pandas as pd
import os
import glob

# 엑셀 파일 읽어오기
excel_file = 'KETI-2017-SL-Annotation-v2_1.xlsx'
df = pd.read_excel(excel_file)

# '번호' 열이 3001~6000 사이인 데이터만 추출
df = df[(df['번호'] >= 3001) & (df['번호'] <= 6000)]

# '한국어' 열을 기준으로 그룹화
grouped = df.groupby('한국어')

# 현재 디렉토리를 비디오 파일이 저장된 폴더 경로로 설정
video_folder = os.getcwd()

# 각 그룹별로 폴더 생성 및 파일 이동
for group_name, group_data in grouped:
    # 데이터 타입 확인 및 변환
    if isinstance(group_name, int):
        output_dir = str(group_name)
    else:
        output_dir = group_name

    # 폴더 생성
    os.makedirs(output_dir, exist_ok=True)

    # 그룹 데이터에서 파일명 추출하여 이동
    for video in group_data['파일명']:
        # 확장자 상관없이 파일 찾기
        pattern = os.path.join(video_folder, f"{os.path.splitext(video)[0]}.*")
        matching_files = glob.glob(pattern)

        # 각 매칭 파일 이동
        for src_file in matching_files:
            dst_file = os.path.join(video_folder, output_dir, os.path.basename(src_file))

            try:
                os.rename(src_file, dst_file)
                print(f"Moved {src_file} to {dst_file}")
            except FileNotFoundError:
                print(f"File {src_file} not found, skipping.")
            except Exception as e:
                print(f"Error moving file {src_file} to {dst_file}: {e}")

```
